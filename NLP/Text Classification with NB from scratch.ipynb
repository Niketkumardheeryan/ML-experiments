```python
def preprocess(l):
    l=[x.lower() for x in l ]
    l=[x.strip('>.":\n<') for x in l]
    l=[x for x in l if x.isalpha()]
    l=[x for x in l if not x in stop]
    return l
```

```python
Y=np.array([i for i in range(20)])
feature_list=vocab
```

```python
def fit(x,y):
    result={}
    result["total_data"]=len(y)
    class_labels=set(y)
    for label in class_labels:
        result[label]={}
        current_rows=(y==label)
        x_current=x[current_rows]
        y_current=y[current_rows]
        total_words=0
        for i in range(len(feature_list)):
            result[label][feature_list[i]]=x_current[:,i].sum()
            total_words+=x_current[:,i].sum()
        result[label]["total_count"]=total_words
    return result
```

```python
def probability(x,dictionary,current_class,k):
    output=np.log(dictionary[current_class]["total_count"])-np.log(dictionary["total_data"])
    for i in range(len(feature_list)):
        word_count=dictionary[current_class][feature_list[i]]+1
        total_count=dictionary[current_class]["total_count"]+(k*len(feature_list))
        word_probability=np.log(word_count+k)-np.log(total_count)
        for j in range(x[i]): # if the frequency of word in test data point is zero then we wont consider it.
            output+=word_probability
    return output
```

```python
def predictSingleClass(x,dictionary,k):
    best_class=-1000
    best_prob=-1000
    firstRun=True
    possible_classes=dictionary.keys()
    logp=[]
    for current_class in possible_classes:
        if current_class=="total_data":
            continue
        class_probability=probability(x,dictionary,current_class,k)
        logp.append(class_probability)
        if(firstRun==True or class_probability>best_prob):
            best_class=current_class
            best_prob=class_probability
        firstRun=False
    return best_class,logp
```

```python
def predict(X_test,dictionary,k):
    num = 0
    Y_pred,logp=predictSingleClass(X_test,dictionary,k)
    return Y_pred,logp
```
