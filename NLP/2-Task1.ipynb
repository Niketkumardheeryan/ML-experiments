## Custom functions for bigrams and trigrams

```python
def bigram(li):
    bi=[]
    for i in range(len(li)-1):
        bi.append((li[i],li[i+1]))
    return bi
        
def trigram(li):
    tri=[]
    for i in range(len(li)-2):
        tri.append((li[i],li[i+1],li[i+2]))
    return tri
```
